```
import requests
form_data = {
    'commandid': 'GetTown',
    'cityid': '01'
}
request_url = "https://emap.pcsc.com.tw/EMapSDK.aspx"
response = requests.post(request_url, data=form_data)
response_content = response.content
print(response_content)
```

```
from lxml import etree
from io import BytesIO
file = BytesIO(response_content)
tree = etree.parse(file)
print(tree)
town_names = [t.text for t in tree.xpath('//TownName')]
print(town_names)
```

```
form_data = {
  'commandid': 'SearchRoad',
  'city': '台北市',
  'town': '松山區',
}
request_url = "https://emap.pcsc.com.tw/EMapSDK.aspx"
response = requests.post(request_url, data=form_data)
response_content = response.content
print(response_content)
file = BytesIO(response_content)
tree = etree.parse(file)
road_names = [t.text for t in tree.xpath('//rd_name_1')]
section_names = [t.text for t in tree.xpath('//section_1')]
print(road_names)
print(section_names)
```

```
form_data = {
  'commandid': 'SearchStore',
  'city': '台北市',
  'town': '松山區',
}
request_url = "https://emap.pcsc.com.tw/EMapSDK.aspx"
response = requests.post(request_url, data=form_data)
response_content = response.content
print(response_content)
file = BytesIO(response_content)
tree = etree.parse(file)
poi_names = [t.text for t in tree.xpath('//POIName')]
print(tree.xpath('//POIName'))
print(poi_names)
```

```
import time
import random
tp_711_stores = {}
for town in town_names:
    form_data = {
      'commandid': 'SearchStore',
      'city': '台北市',
      'town': town,
    }
    request_url = "https://emap.pcsc.com.tw/EMapSDK.aspx"
    response = requests.post(request_url, data=form_data)
    response_content = response.content
    file = BytesIO(response_content)
    tree = etree.parse(file)
    poi_ids = [t.text.strip() for t in tree.xpath('//POIID')]
    poi_names = [t.text for t in tree.xpath('//POIName')]
    lons = [float(t.text)/1000000 for t in tree.xpath('//X')]
    lats = [float(t.text)/1000000 for t in tree.xpath('//Y')]
    adds = [t.text for t in tree.xpath('//Address')]
    tp_711_stores[town] = []
    for poi_id, poi_name, lon, lat, add in zip(poi_ids, poi_names, lons, lats, adds):
        store_info = {
            'POIID' : poi_id,
            'POINAME' : poi_name,
            'LON' : lon,
            'LAT' : lat,
            'ADDRESS' : add
        }
        tp_711_stores[town].append(store_info)
    time.sleep(random.randint(1,6))
    print('Scraping {}'.format(town))


```

```
import requests
request_url = "http://data.nba.net/prod/v2/2019/teams.json"
response = requests.get(request_url)
response_json = response.json()
teams = response_json['league']['standard']

n_nba_teams = 0
for i in teams:
  if i['isNBAFranchise']:
    n_nba_teams += 1

print('2019-2020 球季 NBA 有 {} 支球隊' .format(n_nba_teams))
```

```
team_dict = {}
for t in teams:
  div = t['divName']
  full_name = t['fullName']
  if div in team_dict:
    team_dict[div].append(full_name)
  else:
    team_dict[div] = [full_name]

print(team_dict.keys())
n_ns_teams = len(team_dict['Southeast']) + len(team_dict['Atlantic'])
print('屬於 Atlantic 與 Southwest 的球隊有 {} 個:'.format(n_ns_teams))
print('Atlantic: {}'. format(team_dict['Atlantic']))
print('Southeast: {}'. format(team_dict['Southeast']))
```
